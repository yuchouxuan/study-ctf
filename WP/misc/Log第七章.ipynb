```python
#sove4
import re 
import datetime
with open('log4.log','r') as f:
    logs = f.readlines()

group ={}
for slog in logs:
    ip = re.findall('(.*)  -  -',slog)[0] 
    time = re.findall('\[(.*) \+08',slog)[0] 
    # time = datetime.datetime.strptime(time, '%d/%b/%Y:%H:%M:%S')
    uri = re.findall('POST (.*) HTTP', slog)[0]
    try:
        group[ip].append(uri)
    except:
        group[ip] = [uri]

logCont= {i:[] for i in group }
for i in logCont :
    cont = 0
    for uri in group[i]:
        if 'Login' in uri:
            logCont[i].append(cont)
            cont = 0
        else :
            cont+=1


mapc = 1 
plt.figure(figsize=(20,30))
for i in logCont:
    plt.subplot(9,4,mapc)
    plt.ylim(0,50)
    plt.title(i)
    plt.plot(logCont[i])
    mapc+=1
plt.show()
```
