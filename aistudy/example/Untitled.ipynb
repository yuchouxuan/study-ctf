```python
cols=[0, 1, 2, 3, 5, 6, 7, 10,  12, 13,  15, 20, 21, 22, 23,25, 26, 27, 28, 29]
cols=[0,2,7,20,23,27]
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_iris
from sklearn import svm
from sklearn.feature_selection import VarianceThreshold
torch.cuda.empty_cache()

def boxplot_fill(col):
    # 计算iqr：数据四分之三分位值与四分之一分位值的差
    iqr = col.quantile(0.75)-col.quantile(0.25)
    # 根据iqr计算异常值判断阈值
    u_th = col.quantile(0.75) + 1.5*iqr # 上界
    l_th = col.quantile(0.25) - 1.5*iqr # 下界
    # 定义转换函数：如果数字大于上界则用上界值填充，小于下界则用下界值填充。
    def box_trans(x):
        if x > u_th:
            return u_th
        elif x < l_th:
            return l_th
        else:
            return x
    return col.map(box_trans)

#重新载入数据并进行筛选
iris = load_breast_cancer()
pfs =pd.DataFrame(iris.data)
pfs =(pfs-pfs.min())/(pfs.max()-pfs.min())-0.5
print('原数据集中不同特征的方差：\n', np.var(pfs, axis=0), '\n')
selector = VarianceThreshold(threshold=0.027)
pf= pd.DataFrame(selector.fit_transform(pfs))
cols = pf.columns
in_V=len(cols)

pf['target']=iris['target']
#准备测试-训练数据数据
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(pf[cols],pf['target'],test_size=0.8)
for i in range(in_V):
    x_train_1[i] = boxplot_fill(x_train_1[i])

x_train = torch.tensor(np.array(x_train_1), dtype=torch.float)
x_test = torch.tensor(np.array(x_test_1), dtype=torch.float)
y_train= torch.tensor(np.array(y_train_1), dtype=torch.long)
y_test= torch.tensor(np.array(y_test_1), dtype=torch.long)
x_train_1.max()
```

```python
#训练，预测
clf_linear = svm.SVC(decision_function_shape="ovo", kernel="linear")
clf_rbf = svm.SVC(decision_function_shape="ovo", kernel="rbf")
clf_linear.fit(x_train, y_train)
clf_rbf.fit(x_train, y_train)

y_test_pre_linear = clf_linear.predict(x_test)
y_test_pre_rbf = clf_rbf.predict(x_test)

# 计算分类准确率
from sklearn.metrics import classification_report
print('Linear')
print(classification_report(y_test, y_test_pre_linear))
print('准确率：', accuracy_score(y_test, y_test_pre_linear))
print('pre_rbf')
print(classification_report(y_test, y_test_pre_rbf))
print('准确率：', accuracy_score(y_test, y_test_pre_rbf))
```

```python
import torch.nn.functional as F
torch.cuda.empty_cache()
print(in_V)
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        #输入层
        self.lin = nn.Linear(in_V, in_V*64)
        
        #全连接层
        self.h0= nn.Linear(in_V*64, in_V*64)
        self.h1= nn.Linear(in_V*64, in_V*64)
        self.h2 = nn.Linear(in_V*64, 2)
        
        #输出层
        self.lout = nn.Softmax(dim=1)
        self.drop=nn.Dropout(p=0.2)
        
    def forward(self, x):
        x = F.relu(self.lin(x))
        x=F.relu(self.h0(x))
        x=F.relu(self.h1(x))
        x=F.relu(self.h2(x))
        
        x =F.log_softmax(self.lout(x),1)
        return x



#初始化
net = Net().cuda()# 实例化神经网络
optimizer = optim.Adam(net.parameters(), lr=0.003)  # 据说adam算法比较时髦，且好用
criterion = nn.NLLLoss()#损失函数 负对数

error=[]
zql=[]
for epoch in range(500):
    optimizer.zero_grad()  #梯度清零
    y_pred = net(x_train.cuda())  #反馈
    loss = criterion(y_pred.cuda(), y_train.cuda()) #损失计算
    loss.backward()#
    optimizer.step()
    error.append(loss.item())

    y_pred = net(x_test.cuda())
    y_pred = torch.argmax(y_pred.cuda(), dim=1)
    
    y_test=y_test.cpu()
    y_pred=y_pred.cpu()
    zql.append(float(accuracy_score(y_test, y_pred)))
    

import datetime 
torch.save(net.state_dict,f"c://CTF//123")
y_test=y_test.cpu()
y_pred=y_pred.cpu()
# it is necessary that drawing the loss plot when we fine tuning the model
sns.lineplot(data=error)
sns.lineplot(data=zql)
plt.show()
y_test=y_test.cpu()
y_pred=y_pred.cpu()
print(classification_report(y_test, y_pred))
print('准确率：', accuracy_score(y_test, y_pred))
torch.cuda.empty_cache()
torch.cuda.device_count()
```
