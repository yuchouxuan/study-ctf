```python
import cv2
import numpy as np
im = cv2.imread('lena.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
# plt.imshow(im)
```

# 翻转

```python
plt.subplot(1,3,1)
plt.imshow(im)

#水平
im_f1ip0 = cv2.flip(im, 0)
plt.subplot(1,3,3)
plt.imshow(im_f1ip0)

#垂直
im_f1ip1 = cv2.flip(im, 1)
plt.subplot(1,3,2)
plt.imshow(im_f1ip1)
```

# 旋转

```python
def rotate(img, angle, center=None, scale=1.0):
    # 获取图像尺寸
    (h, w) = img.shape[:2]

    # 旋转中心的缺失值为图像中心
    if center is None:
        center = (w / 2, h / 2)

    # 调用计算旋转矩阵函数

    M = cv2.getRotationMatrix2D(center, angle, scale)

    # 使用Opencv仿射变换实现旋转操作
    rotated = cv2.warpAffine(img, M, (w, h))

    # 返回旋转后的图像
    return rotated

# 加载图像并显示
plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
plt.imshow(im)

# 对原图做旋转操作
# 逆时针50度
rotated = rotate(im, 45)
plt.subplot(1,4,2)
plt.imshow(rotated)
# 顺时针20度
rotated = rotate(im, -20)
plt.subplot(1,4,3)
plt.imshow(rotated)
# 逆时针90度
rotated = rotate(im, 90)
plt.subplot(1,4,4)
plt.imshow(rotated)
```

# 平移

```python
def translate(img, x, y):
    # 获取尺寸
    (h, w) = img.shape[:2]

    # 定义平移矩阵
    M = np.float32([[1, 0, x], [0, 1, y]])

    # 使用OpenCv仿射变换函数实现平移操作
    shifted =cv2.warpAffine(img, M, (w, h))

    #返回转换后的图像
    return shifted

# 加载图像并显示
plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
plt.imshow(im)

# 对原图做平移操作
# 下移50像素
shifted = translate(im, 0, 50)
plt.subplot(1,4,2)
plt.imshow(shifted)
# 左移100像素
shifted = translate(im, -100, 0)
plt.subplot(1,4,3)
plt.imshow(shifted)
# 右移50，下移100像素
shifted = translate(im, 50, 100)
plt.subplot(1,4,4)
plt.imshow(shifted)
```

# 缩放

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.imshow(im)

#获取图像尺寸
(h, W) = im.shape[:2]
#缩放的目标尺寸
dst_size = (200, 300)
#最邻近插值
method = cv2.INTER_NEAREST
#进行缩放
resized = cv2.resize(im, dst_size, interpolation = method)
plt.subplot(1,3,2)
plt.imshow(resized)
#缩放的目标尺寸
dst_size = (800, 600)
#双线性插值
method = cv2.INTER_LINEAR
#进行缩放
resized = cv2.resize(im, dst_size, interpolation = method)
plt.subplot(1,3,3)
plt.imshow(resized)
```
